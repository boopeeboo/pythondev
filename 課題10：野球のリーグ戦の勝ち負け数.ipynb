```python
class Baseball_Team:
    def __init__(self, name, win,lose,draw):
        self.name = name
        self.win = win
        self.lose = lose
        self.draw = draw

    def calc_win_rate(self):
        total_games = self.win + self.lose
        if total_games == 0:
            return 0.0
        return self.win / total_games
        
    def show_team_result(self):
        win_rate = self.calc_win_rate()
        print(f"{self.name:10} {self.win:5} {self.lose:7} {self.draw:6} {win_rate:.3f}")

```

```python
Archers = Baseball_Team("Archers", 77, 64, 2)
Meteors = Baseball_Team("Meteors", 71, 69, 3)
Cheetahs = Baseball_Team("Cheetahs", 69, 68, 6)
Sharks = Baseball_Team("Sharks", 70, 70, 3)
Pythons = Baseball_Team("Pythons", 68, 73, 2)
Owls = Baseball_Team("Owls", 59, 82, 2)

teams = [Archers, Meteors, Cheetahs, Sharks, Pythons, Owls]

print(f"{'Team':10} {'Wins':5} {'Losses':7} {'Draws':6} {'WinRate'}")
print("-" * 40)

for t in teams:
    print(f"{t.name:10} {t.win:5} {t.lose:7} {t.draw:6} {t.calc_win_rate():.3f}")


```

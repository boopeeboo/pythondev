```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([62, 95, 83, 100, 72, 71, 55, 85, 67, 75])   # english
y = np.array([62, 80, 85, 95, 58, 76, 52, 84, 67, 71])    # math
```

```python
p = np.polyfit(x, y, 1)
f = np.poly1d(p)
plt.xlabel("english")
plt.ylabel("math")
plt.scatter(x, y, label="score", color="blue") #散布図は青色
plt.plot(x, f(x), label="lsm", color="red")  #赤い線で直線を描く)
plt.legend(loc="upper left")

plt.show()
```

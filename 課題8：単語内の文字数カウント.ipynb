```python
c = []
c = dict.fromkeys(list("abcdefghijklmnopqrstuvwxyz"), 0)

word = []

while True:
    t1 = input("英単語を入力してください： ")

    if t1 == "":
        break
    word.append(t1)
    print(f"英単語を入力してください: {t1}")
    for t2 in t1:
        if t2 in c:
            c[t2] += 1

word.sort()
print(f"入力した英単語： {word}")

for key, value in c.items():
    if value > 0:
        print(f"{key} が {value} 個ありました")
```

```python
try:
    a = input("割られる数を入力してください：")
    a = int(a)
    b = input("割る数を入力してください：")
    b = int(b)
    
except:
    print("割られる数を入力してください：" + str(a))
    print("エラー: 数値を入力してください")
    
else:
    try:
        c = a / b
    except ZeroDivisionError:
        print("割られる数を入力してください：" + str(a))
        print("割る数を入力してください：" + str(b))
        print("エラー:0で割り算しないでください")
    else:
        print("割られる数を入力してください：" + str(a))
        print("割る数を入力してください：" + str(b))
        print(f"{a} ÷ {b} = {c}")
    
finally:
    print("処理を終了します")
```
